### 6. Cross tabulations

```python
df6= df2
# # mean = df5['Total'].mean()
```

```python
df6['tot_range'] = 'Low'
df6['ren_range'] = 'low'
df6['non-ren_range'] = 'lo'
```

```python
df6['tot_range'][df6['Total']>=df6['Total'].mean()] = 'High'
df6['ren_range'][df6['renewable']>=df6['renewable'].mean()] = 'high'
df6['non-ren_range'][df6['non-renewable']>=df6['non-renewable'].mean()] = 'hi'
```

```python
df6
```

```python
month = np.array(df6['month'])
year = np.array(df6['year'])
tot_range = np.array(df6['tot_range'])
ren_range = np.array(df6['ren_range'])
non_ren_range = np.array(df6['non-ren_range'])
```

```python
pd.crosstab(month, [tot_range, ren_range,non_ren_range], rownames=["month"], colnames=["tot_range", "ren_range","non-ren_range"])
```

```python
pd.crosstab([month,tot_range], [ren_range,non_ren_range], rownames=["month",'tot_range'], colnames=[ "ren_range","non-ren_range"])
```

```python
pd.crosstab(df6['month'],df6['tot_range']) 
#If crosstab() receives only two Series, it will provide a frequency table.
```

```python
# crosstab() can also be implemented to Categorical data.
pd.crosstab(ren_range,non_ren_range)
```

### 6.1 Normalization

###### Frequency tables can also be normalized to show percentages rather than counts using the normalize argument:

```python
pd.crosstab(df6['month'],df6['tot_range'], normalize=True)
```

###### can also normalize values within each row or column:

```python
pd.crosstab(df6['month'],df6['tot_range'], normalize='columns')
```

```python
pd.crosstab(df6['month'],df6['tot_range'], normalize='index')
```

###### can also be passed a third Series and an aggregation function (aggfunc) that will be applied to the values of the third Series within each group defined by the first two Series

```python
pd.crosstab(df6['month'],df6['year'],values=df6['Total'],aggfunc=np.sum)
```

### 6.2 Adding margins

###### Finally, one can also add margins or normalize this output.

```python
pd.crosstab(df6['month'],df6['year'],values=df6['Total'],aggfunc=np.sum,normalize=True,margins=True)
```
